```python
from eBallena import eBallena
from Baseline import simulacionBrian2Voltages

import numpy as np
import matplotlib.pyplot as plt
```

```python
# =========================
#     RANDOM INSTANCE AND
#        SYNAPSES
# =========================

# configuracion de la simulacion
config   = eBallena.createConfig(n_inputs=2,n_neu=2,tau=10,max_sim_time=100, refractory_time=1)

# definir inputs aletarios
n_spikes = np.random.randint(4,10)
times    = np.random.choice(np.arange(10, 90), size=n_spikes, replace=False)
neu_idx  = np.random.randint(0,2, size=n_spikes)
instance = [ e for e in zip(times,neu_idx) ]

# definir persos sinapticos aleatorios
w       = np.random.randint(10,100, size=4)
syn_in   = [(0,0,w[0]),(1,1,w[1])]
syn_re   = [(0,1,w[2]),(1,0,w[3])]

# =========================
#    HACER SIMULACIONES
# =========================
res = eBallena.simulate(config, instance, syn_in, syn_re, mode='VOLTAGES')
time, voltages = simulacionBrian2Voltages(config, instance, syn_in, syn_re)



# =========================
#    PLOTEAR
# =========================
plt.figure( figsize=(15,5) )

plt.subplot(141)
plt.title('Neu 0')
plt.plot(time,voltages[0], label='Brian2')
eBallena.plotVoltage(voltage_markers=res,neu=0, config=config, color='#F55718')
plt.grid()
plt.legend()

plt.subplot(142)
plt.title('Neu 1')
plt.plot(time,voltages[1], label='Brian2')
eBallena.plotVoltage(voltage_markers=res,neu=1, config=config, color='#F55718')
plt.grid()
plt.legend()

plt.subplot(143)
plt.title('Neu defasado')
plt.plot(time-5,voltages[0], label='Brian2')
eBallena.plotVoltage(voltage_markers=res,neu=0, config=config, color='#F55718')
plt.grid()
plt.legend()

plt.subplot(144)
plt.title('Neu 1 defasado')
plt.plot(time-5,voltages[1], label='Brian2')
eBallena.plotVoltage(voltage_markers=res,neu=1, config=config, color='#F55718')
plt.grid()
plt.legend()


plt.show()
```
